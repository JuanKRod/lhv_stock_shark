```python
DIRS = os.getcwd()
all_files = []
for root, folder, files in os.walk(DIRS):
    all_files.append(files)

all_files=all_files[0]

complete_stock_file =[]
for file in all_files:
    if '.csv' in file:
        if 'ticker_list.csv' in file:
            pass
        else:
            complete_stock_file.append(file)
```

```python
# RUN ONLY 1 TIME 
# THIS ADDS A NEW COLUMN AND ADDS THE TICKER SYMBOL 

for file in complete_stock_file:
    data = pd.read_csv(file)
    ticker_value = file.split('.')[0]
    data['ticker']= ticker_value
    data.to_csv(file)
```

```python
# RUN ONLY 1 TIME
# TO REMOVE COLUMNS

for file in complete_stock_file:
    data = pd.read_csv(file)
    col = data.columns
    data=data.drop(col,axis=1)
    data.to_csv(file)
```
